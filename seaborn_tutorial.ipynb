## Seaborn Tutorial 정리 

## 시각화와 통계의 관계
  
 - 변수간의 관계,영향을 이해하는 것이 통계적 것이 통계적 분석
 - replot()
    

  - replot(): 의미단위의 두변수를 각각의 다른 축으로 관계를 보여줌
     - sns.replot(x= '독립변수',y= '종속변수', data= '데이터 명,hue or style, , ,추가, ' kind= 'line' or 'scatter')
   - 예시
  - scatterplot()  =>default

```
#seaborn을 활용하기위한 패키지 실행

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

## 산점도(scatter plot) 그리기

- 산점도는 통계시각화의 핵심
- 데이터셋에서 관측된 두 변수의 개별 데이터를 점을 이용해서 표현
- 두 변수사이 유의미한 관계 존재여부를 알려줌   

```
tips = sns.load_dataset("tips")                   # 내장데이터 불러오기, 이름짓기
sns.relplot(x="total_bill", y="tip", data=tips);  # 변경한 데이터셋이름으로 산점도 그리고 독립변수,종속변수설정
```

- 색상을 이용해서 또 다른 변수를 추가할 수 있음(hue semantic)


```
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips); #data 입력하기전에 변수 추가/독,종 변수 다음
```

- 변수간 차이를 더 보여주고 싶을때는 표시점의 모양에 변화를 줄 수 있음

```
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips);  # hue다음에 style항목 기입
```

- 색상변화와 표시점 모양을 모두 변형해서 4개의 변수를 보여줄수있음
- 주의점: 색상변화가 모양변화보다 더 눈에 띔

```
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);
```

- 색상으로 변화를 줄 변수가 수치, 정수로 이루어진 범주형 변수면 qualititive palette가 자동적으로 적용
- 색상의 연속적 변화를 통해 변수의 증감 및 차이를 보여줌
- 사용자가 원하는 색상으로 바꾸거나 marker의 크기 변화가능(size 차이에 따른 크기 변화, 전체 marker의 크기 변화 모두 가능)

```
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

```
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips);
```

```
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
```

```
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);
```

## 선표본점(line plot) 그리기

- 선표본점으로 연속성 강조하기

- 산점도 뿐만 아니라 시간의 흐름에 따른 변화나 연속성이 있는 변수등을 표현할 경우 선을 이용해서 선표본점을 그릴수 있음

- sns.replot( , , , , kind='line', , )

```
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
# numpy를 이용해서 딕셔너리 형태로 'time','value' 키값을 만들고 
# 밸류값을 time은 0~500범위,value는 0~500사이의 범위의 랜덤값을 누적합해서
# 딕셔너리타입의 numpy자료를 만든후 판다스를 이용해서 데이터프레임으로 만들고 이름을 지음 
g = sns.relplot(x="time", y="value", kind="line", data=df)
# 데이터프레임으로 만든 자료를 시각화하기위해 seaborn을 이용한 코드입력
# 독립변수와 종속변수(의미를 찾고자 하는 변수)를 지정하고 
# 선표본점은 디폴트가 아니기때문에 선표본점을 사용한다는 코드, 데이터프레임 이름을 입력후
# 시각화 자료이름을 새롭게 부여
g.fig.autofmt_xdate()
```

```
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
sns.relplot(x="x", y="y", sort=False, kind="line", data=df)
```
